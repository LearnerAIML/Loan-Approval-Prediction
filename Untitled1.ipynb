```python
import joblib
import pandas as pd

model = joblib.load('loan_approval_model.pkl')

no_of_dependents = 2
education = 1
self_employed = 0
income_annum = 5000000
loan_amount = 10000000
loan_term = 12
cibil_score = 700
residential_assets_value = 3000000
commercial_assets_value = 2000000
luxury_assets_value = 1000000
bank_asset_value = 500000

input_data = pd.DataFrame([[no_of_dependents, education, self_employed,
                            income_annum, loan_amount, loan_term, cibil_score,
                            residential_assets_value, commercial_assets_value,
                            luxury_assets_value, bank_asset_value]],
                          columns=['no_of_dependents', 'education', 'self_employed',
                                   'income_annum', 'loan_amount', 'loan_term', 'cibil_score',
                                   'residential_assets_value', 'commercial_assets_value',
                                   'luxury_assets_value', 'bank_asset_value'])

result = model.predict(input_data)[0]
print("Loan Status:", "Approved ✅" if result == 1 else "Rejected ❌")
```
